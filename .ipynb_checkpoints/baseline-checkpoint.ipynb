```python
import os, sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

sys.path.append(os.path.join(os.path.abspath(".."), "code"))
```

```python
# Attribution: [code based on pytorch docs](https://pytorch.org/tutorials/beginner/transfer_learning_tutorial.html?highlight=transfer%20learning)

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
from torch.optim import lr_scheduler
from torchvision import datasets, models, transforms, utils

DATA_DIR = os.path.join(os.path.abspath(".."), "cons454_project/data/")
IMAGE_LENGTH = 711
IMAGE_WIDTH = 400

data_transforms_bw = {
    "train": transforms.Compose(
        [
            transforms.Resize((IMAGE_WIDTH, IMAGE_LENGTH)), 
            transforms.Grayscale(num_output_channels=1),            
            transforms.ToTensor(),
        ]
    ),
    "valid": transforms.Compose(
        [
            transforms.Resize((IMAGE_WIDTH, IMAGE_LENGTH)),                        
            transforms.Grayscale(num_output_channels=1),            
            transforms.ToTensor(),
        ]
    ),
}
image_datasets_bw = {
    x: datasets.ImageFolder(os.path.join(DATA_DIR, x), data_transforms_bw[x])
    for x in ["train", "valid"]
}
dataloaders_bw = {
    x: torch.utils.data.DataLoader(
        image_datasets_bw[x], batch_size=24, shuffle=True, num_workers=4
    )
    for x in ["train", "valid"]
}
dataset_sizes = {x: len(image_datasets_bw[x]) for x in ["train", "valid"]}
class_names = image_datasets_bw["train"].classes
```

```python
# Get a batch of training data
inputs_bw, classes = next(iter(dataloaders_bw["train"]))
```

```python
plt.figure(figsize=(10, 8)); plt.axis("off"); plt.title("Sample Training Images")
plt.imshow(np.transpose(utils.make_grid(inputs_bw, padding=1, normalize=True),(1, 2, 0)));
```

```python
flatten_transforms = transforms.Compose([
                    transforms.Resize((IMAGE_WIDTH, IMAGE_LENGTH)),
                    transforms.Grayscale(num_output_channels=1),
                    transforms.ToTensor(),
                    transforms.Lambda(torch.flatten)])
train_flatten = torchvision.datasets.ImageFolder(root=DATA_DIR + 'train', transform=flatten_transforms)
valid_flatten = torchvision.datasets.ImageFolder(root=DATA_DIR + 'valid', transform=flatten_transforms)
```

```python
train_dataloader = torch.utils.data.DataLoader(train_flatten, batch_size=100, shuffle=True)
valid_dataloader = torch.utils.data.DataLoader(valid_flatten, batch_size=50, shuffle=True)

flatten_train, y_train = next(iter(train_dataloader))
flatten_valid, y_valid = next(iter(valid_dataloader))
```

```python
dummy = DummyClassifier()
dummy.fit(flatten_train.numpy(), y_train)
dummy.score(flatten_train.numpy(), y_train.numpy())
```

```python
dummy.score(flatten_valid.numpy(), y_valid.numpy())
```

```python
lr_flatten_pipe = make_pipeline(StandardScaler(), LogisticRegression(max_iter=3000))
lr_flatten_pipe.fit(flatten_train.numpy(), y_train)
```

```python
lr_flatten_pipe.score(flatten_train.numpy(), y_train.numpy())
```

```python
lr_flatten_pipe.score(flatten_valid.numpy(), y_valid.numpy())
```
